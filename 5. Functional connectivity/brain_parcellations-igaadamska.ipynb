# Plotting brain parcellations

## Load libraries

```python
from nilearn import datasets
from nilearn import plotting
import numpy as np
```

## Plot brain atlas from Nifti file

```python
# Load atlas
ho_atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-2mm')
ho_maps = ho_atlas.maps

# Plot atlas
plotting.plot_roi(ho_maps, title="Harvard Oxford atlas")
plotting.show()
```

## Plot brain atlas created based on MNI coordinates

```python
# Load atlas
power = datasets.fetch_coords_power_2011()

# Select coordinates
coords = np.vstack((power.rois['x'], power.rois['y'], power.rois['z'])).T

# Print coordinates
print(coords)
```

```python
# Plot on brain slices

display = plotting.plot_roi(None)
display.add_markers(coords, marker_size=100)
```

```python
# Plot on glass brain
n_rois = len(coords)
connectivity_matrix = np.zeros((n_rois, n_rois))

# Plot using plot_connectome function

plotting.plot_connectome(connectivity_matrix,coords)
plt.show()
```

```python
# Load libraries

import nibabel as nib
from nilearn import plotting, image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data

fmri_path = 'data/sub-01_ses-1_task-rest_bold_space-MNI152NLin2009cAsym_preproc.nii.gz'
fmri = image.load_img(fmri_path)
```

```python
# Print dimensions and type of data

fmri.shape
```

```python
# Select one volume (image.index_img) 

fmri_1 = image.index_img(fmri, 30)
fmri_1.shape
```

```python
# Make a plot (plotting.plot_epi)

plotting.plot_epi(fmri_1)
```

```python
# Convert nibabel file to numpy array (get_data)
fmri_array = fmri.get_data()

# Check type and dimensions
fmri_array.shape
```

```python
# Plot timeseries of selected voxels
voxel_1 = fmri_array[49,20,39,:]
voxel_2 = fmri_array[60,26,35,:]

plt.plot(voxel_1)
plt.plot(voxel_2)
plt.legend(["voxel 1", "voxel 2"])
```

```python
# Generate random timeseries with the length of fMRI data (np.random.uniform (low, high, size=(,))

random = np.random.uniform(voxel_1.min(), voxel_1.max(), size=len(voxel_1))
```

```python
# Plot random timeseries with selected voxel timeseries

plt.plot(voxel_1)
plt.plot(random)
plt.legend(["Real timeseries", "Random timeseries"])
```
